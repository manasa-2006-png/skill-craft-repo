```python
import pandas as pd
import matplotlib.pyplot as plt

# Read dataset
df = pd.read_csv("population_set.csv")   # or "population_set.csv"

# Define age bins and labels
bins = [0, 20, 40, 60, 80]
labels = ["0-20 Years", "21-40 Years", "41-60 Years", "61-80 Years"]

# Create a new column for Age Ranges
df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels, include_lowest=True)

# Count population per age group
age_counts = df["AgeGroup"].value_counts().sort_index()

# Colors for infographic style
colors = ["#FFD700", "#1E90FF", "#FF69B4", "#32CD32"]

# Plot
plt.figure(figsize=(9,6))
bars = plt.bar(age_counts.index.astype(str), age_counts.values, color=colors, edgecolor="black")

# Add labels inside bars
for bar, count in zip(bars, age_counts.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height()/2,
             str(count), ha="center", va="center", fontsize=12, fontweight="bold", color="white")

# Titles and labels
plt.title("Population Distribution by Age Groups", fontsize=16, fontweight="bold")
plt.xlabel("Age Groups", fontsize=13)
plt.ylabel("Number of People", fontsize=13)

# Clean infographic look
plt.grid(False)
plt.gca().set_facecolor("white")

plt.tight_layout()
plt.show()
```
